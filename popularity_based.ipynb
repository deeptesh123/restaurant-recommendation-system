```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('always')
warnings.filterwarnings('ignore')
```

```python
zomato=pd.read_csv("../zomato.csv")
zomato.columns
```

```python
pop=zomato[["name","rate","votes","approx_cost(for two people)"]]
pop["rate"]=pop.rate.str.replace("/5","")
pop=pop[pop["rate"]!="NEW"]
pop=pop[pop["rate"]!="-"]
pop["rate"]=pop["rate"].str.strip("")
pop["rate"]=pop.rate.astype(float)
pop.rename(columns={"approx_cost(for two people)":"cost"},inplace=True)
pop["cost"]=pop.cost.str.replace(",","")
pop["cost"]=pop.cost.astype(float)
pop
```

```python
legit=pop.groupby("name").mean()[["votes","rate","cost"]].reset_index()

popular=legit[legit["votes"]>500].sort_values("rate",ascending=False).head(10)
popular
```

```python
popular.head(2)
```

```python
cuisines=[]
for name in popular["name"]:
    idx=zomato[zomato["name"]==name].index[0]
    cuisines.append(zomato.loc[idx,"cuisines"])
popular["cuisines"]=cuisines
popular["cost"]=popular.cost.apply(lambda x:round(x))
popular["rate"]=popular.rate.apply(lambda x:round(x,2))
popular
```

```python
for i in range(10):
    print(popular.iloc[i]["name"],popular.iloc[i]["rate"],popular.iloc[i]["cuisines"],popular.iloc[i]["cost"])
```

```python
popular.to_csv("popular.csv")
```
